## What is the value of the first triangle number to have over five hundred divisors?

**Solution(s):** 

We'll define a function *divisorGenerator* that takes in $n$ and gives us the divisors of $n$. We'll then call the on the triangular numbers until we have enough divisors. The formula for triangular numbers $m(m+1)/2$ will be used instead of computing the sum.

```python
from math import sqrt
def divisorGenerator(n):
    # keeping two arrays so that once we have a divisor d, we know that n/d is also a divisor. 
    small_divisors = []
    large_divisors = []
    for i in range(1, int(sqrt(n) + 1)):
        if n % i == 0:
            small_divisors.append(i)
            if i**2 != n:
                large_divisors.append(int(n / i))
    small_divisors.extend(reversed(large_divisors))
    return small_divisors
```

```python
# run a while loop that exits when the list of divisors is long enough
i = 1
while len(divisorGenerator(i*(i+1)/2)) <= 500:
    i += 1
print(i*(i+1)/2)
```
