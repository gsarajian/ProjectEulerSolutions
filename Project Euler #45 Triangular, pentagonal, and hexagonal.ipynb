## Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

## Triangle	 	$T_n=n(n+1)/2$,	 	$1, 3, 6, 10, 15, ...$
## Pentagonal	 	$P_n=n(3n−1)/2$,	 	$1, 5, 12, 22, 35, ...$
## Hexagonal	 	$H_n=n(2n−1)$,	 	$1, 6, 15, 28, 45, ...$
It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.

**Solution(s):**


We generate a list of all three types of numbers and keep track of three indices which indicate where in the lists we are. We compare if they're all equal. If not, we take the index of the smallest of the three and increase it. 

```python
# Generate the lists of numbers
tris = [n*(n+1)/2 for n in range(2,10**5)];
pents = [n*(3*n-1)/2 for n in range(2,10**5)];
hexs = [n*(2*n-1)for n in range(2,10**5)];

# introduce indices and a boolean to help us get past the first time they're the same
i,j,k = 0,0,0
found = False

while i < 99998 and j < 100000 and k < 100000:
    tri = tris[i]
    pent = pents[j]
    hecks = hexs[k]
    if tri == pent and pent == hecks:
        if not found:
            print("we found one", pent, i, j, k)
            found = True
        else:
            print("Found second", pent, i, j, k)
            break
    if tri == min(tri, pent, hecks):
        i += 1
    elif pent == min(tri, pent, hecks):
        j += 1
    elif hecks == min(tri, pent, hecks):
        k += 1
```
