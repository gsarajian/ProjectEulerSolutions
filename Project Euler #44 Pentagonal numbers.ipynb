##  Pentagonal numbers are generated by the formula, $P_n=n(3n−1)/2$. The first ten pentagonal numbers are:

$$1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...$$

## It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, $70 − 22 = 48$, is not pentagonal.

## Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k − P_j|$ is minimised; what is the value of $D$?

**Solution(s):**
We simply brute force, running through all pairs of the first 10,000 pentagonal numbers and checking if their difference and sum are both pentagonal. There's only one such example in this window, and we're lucky that it works. To prove that it's indeed the smallest, we'd have to continue searching through about the first 2.3 million pentagonal numbers. I'm not sure we're supposed to actually show our answer is correct, Idk how this is a 5% problem.

```python
pents = [n*(3*n-1)//2 for n in range(1,10**8)];
hsh = set(pents)
```

```python
pents2 = [n*(3*n-1)//2 for n in range(10**5)];
hsh2 = set(pents2)
```

```python
cands = []
for i in range(1,10**4):
    b = pents2[i]
    for j in range(i,10**4):
        c = pents2[j]
        if c + b in hsh and c-b in hsh:
            cands.append([b, c])
```
